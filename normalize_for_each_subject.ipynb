```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import collections
import pprint

path_hue = "./data/pilot_study/csv/each_parameter/df_each_parameter_hue.csv"
out_path_subject = "./data/pilot_study/csv/each_subject/"
out_path_overall = "./data/pilot_study/csv/each_parameter/"
# initialization
each_parameter_list_hue = pd.read_csv(path_hue, header=0, index_col=0)
index_list = each_parameter_list_hue.index.values
columns_list = each_parameter_list_hue.columns.values
df_subject_list = []
df_overall = pd.DataFrame()
for subject in range(12):
    df_subject_list.append(pd.DataFrame(index=index_list,
                                        columns=["order_" + str(i) for i in range(17)]))
    
# process each order
for subject in range(12):
    for order in range(17):
        df_subject_list[subject]["order_" + str(order)] = each_parameter_list_hue["order_" + str(order) + "-" + str(subject)]
        
for subject in range(12):
    for i, il in enumerate(index_list):
        hoge = (df_subject_list[subject].loc[il] - df_subject_list[subject].loc[il].min()) / (df_subject_list[subject].loc[il].max() - df_subject_list[subject].loc[il].min())
        df_subject_list[subject].loc[il] = hoge
    df_subject_list[subject].to_csv(out_path_subject + "normalized_subject_" + str(subject) + ".csv")

# concat all subject's normalized parameter
df_overall = pd.concat(df_subject_list, axis=1)
df_overall.to_csv(out_path_overall + "df_normalized_each_subject_parameter.csv")
```
