```python
import pandas as pd
from pandas.io.json import json_normalize
import matplotlib as plt
import pprint
import json
```

```python
subject_list = ["okamoto", "kataoka",
                "ikari", "tanaka", 
                "huruki", "omura", 
                "haga", "kawakata", 
                "kobayashi", "oe", 
                "kato"]
parameter = ["valence", "arousal",
            "userValence", "userArousal"]
df_list = []
df_param_list = []
df_overall = None
order_num = 17
```

```python
# read JSONs and convert them into DataFrame
for i, sl in enumerate(subject_list):
    with open("./data/experiment/json/" + sl + "_experiment.json") as f:
        json_each_subject = json.load(f)
        # delete unused parameter
        print(sl)
        del json_each_subject["is_pilot_experiment"], json_each_subject["is_experiment"], json_each_subject["order"]
        for j in range(order_num):
            print(json_normalize(json_each_subject[str(j)]).reset_index(drop=True))
            df_list.append(json_normalize(json_each_subject[str(j)]))
```

```python
df_overall = pd.concat(df_list).reset_index(drop=True)
# save as xlsx
with pd.ExcelWriter("./data/experiment/csv/overall.xlsx") as w:
    df_overall.to_excel(w)
# save as csv
df_overall.to_csv("./data/experiment/csv/overall.csv", index=False)
```
